```python
import pandas as pd
import os
```

```python
books = []
for file in os.listdir('.'):
    if 'xlsx' in file:
        books.append(file)
      
```

```python
books1 = pd.read_excel(books[0], sheet_name=None, index_col=False)
books2 = pd.read_excel(books[1], sheet_name=None, index_col=False)
```

```python
result_dct = {}
for key2 in books2:
    if key2 not in books1.keys():
        result_dct.setdefault(key2, 'Нет такой вкладки')
    else:
        result_dct.setdefault(key2, '')
        d2 = books2[key2] 
        d1 = books1[key2]
        d1 = d1.fillna('')
        d2 = d2.fillna('')
        d2.columns = d2.columns.str.strip()
        d1.columns = d2.columns.str.strip()
        try:
            d3 = d2 == d1
            if d3.all().all():
                result_dct[key2] = 'Вкладки совпадают'
            else:
                temp_arr = []
                for index in d3.index:
                    for i, col in enumerate(d3.columns):
                        temp_cell = d3.iloc[index, i]
                        if not temp_cell:
                            temp_arr.append((f'Строка {index+2}', f'Колонка {i+1}'))
                result_dct[key2] = temp_arr            
        except:
            result_dct[key2] = 'Несовпадение по кол-ву столбцов или строк'
            continue
```
