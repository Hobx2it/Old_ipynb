```python
from urllib.request import urlopen, urlretrieve
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import re
import pandas as pd
from datetime import datetime
import time
import requests 
from fake_headers import Headers 
import os
```

```python
urls = pd.read_excel(r'C:\Users\AmanovRA\Текущее\Логопедические центры\Ссылки_ZOON.xlsx')['Ссылки'].to_list()

logoped_centrs = {'name' : [],
                 'web_site' : [],
                 'zoon_url' : []}

for url in urls:
    logoped_centrs['zoon_url'].append(url)
    resp = urlopen(url)
    html = resp.read().decode('utf8') # считываем содержимое
    soup = BeautifulSoup(html, 'html.parser') # делаем суп

    for title in soup.find_all('title'):
        logoped_centrs['name'].append(title.text[:title.text.find(' - отзывы')])
        cnt = 0

    for dt in soup.find_all('dt'):
        if dt.text == 'Официальный сайт':
            link = dt.find_next('a')
            logoped_centrs['web_site'].append(link.text)
            cnt += 1
    if not cnt:
        logoped_centrs['web_site'].append('')
            
lc = pd.DataFrame(logoped_centrs)            
```

```python
lc.to_excel(r'C:\Users\AmanovRA\Текущее\Логопедические центры\Сайты_центров.xlsx')
```
