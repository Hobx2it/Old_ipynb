```python
import pandas as pd
import datetime
```

```python
base = pd.read_excel('СПУ_1.xlsx')
```

```python
base = base.append(pd.read_excel('СПУ_2.xlsx'))
base
```

```python
base = base.reset_index(drop=True)
```

```python
def get_date(x):
    try:
        x = x.date()
        return x
    except:
        return x
    

base['Дата окончания СПУ из карточки Пациента'] = base['Дата окончания СПУ из карточки Пациента'].apply(get_date)
```

```python
base
```

```python
base_end = base[base['Дата окончания СПУ из карточки Пациента'] == '-']
```

```python
base_end
```

```python
temp = base[base['Дата окончания СПУ из карточки Пациента'] != '-']
```

```python
base_end = base_end.append(temp[temp['Дата окончания СПУ из карточки Пациента'] >= datetime.date(2020,9,24)])
```

```python
base_end = base_end.reset_index(drop=True)
```

```python
base_end['Дата начала СПУ из карточки Пациента']  = base_end['Дата начала СПУ из карточки Пациента'].apply(get_date)
```

```python
base_end_new = base_end[base_end['Дата начала СПУ из карточки Пациента'] <= datetime.date(2020,9,24)]
```

```python
group_base_end_new = base_end_new.groupby('Номер карты', as_index=False).agg({'СПУ из карточки Пациента':list})
```

```python
group_base_end_new
```

```python
data = pd.read_excel('Ошибки ПППК на 21.09.2020.xlsx')
```

```python
data['Номер карты'].nunique()
```

```python
data = data.merge(group_base_end_new, how='left', left_on='Номер карты', right_on='Номер карты')
```
