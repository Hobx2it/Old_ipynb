```python
import pandas as pd
from difflib import get_close_matches
```

```python
nsk = pd.read_excel('Лабораторная диагностика.xls')
spb = pd.read_excel('Копия лаб_СПБ_Настя.xls', dtype={'Код' : str})
```

```python
nsk_arr = nsk['Наименование услуги'].to_list()
spb_arr = spb['Наименование'].to_list()
```

```python
nsk_arr = {i : '' for i in nsk_arr}
```

```python
for key in nsk_arr:
    nsk_arr[key] = get_close_matches(key, spb_arr, n=2, cutoff=0.5)
```

```python
nsk_arr
```

```python
for key, val in nsk_arr.items():
    if len(val) == 2:
        continue
    else:
        nsk_arr[key] += [''] * (2 - len(val))
```

```python
data = pd.DataFrame(nsk_arr)
```

```python
data = data.T
```

```python
data = data.rename(columns={0:'Первое сопоставление СПБ', 1:'Второе сопоставление СПБ'})
```

```python
data['База НСК'] = data.index

```

```python
data.index = range(data.shape[0])
```

```python
data
```

```python
spb_dict = dict(zip(spb['Наименование'].to_list(), spb['Код'].to_list()))
```

```python
spb_dict
```

```python
temp_arr = []
for word in data['Первое сопоставление СПБ']:
    if word in spb_dict:
        temp_arr.append(spb_dict[word])
    else:
        temp_arr.append('')
```

```python
temp_arr2 = []
for word in data['Второе сопоставление СПБ']:
    if word in spb_dict:
        temp_arr2.append(spb_dict[word])
    else:
        temp_arr2.append('')
```

```python
data['Код первый'] = temp_arr
data['Код второй'] = temp_arr2
```

```python
data
```

```python
data.to_excel('Сопоставления.xlsx')
```
