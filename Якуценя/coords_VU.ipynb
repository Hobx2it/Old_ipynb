```python
import pandas as pd
```

```python
data = pd.read_excel('coords_adresses.xlsx')
```

```python
coords = data[['Координаты']]
```

```python
coords.dropna(inplace=True)
```

```python
coords.Coord = coords.Coord.apply(lambda x: list(map(float, x.split())))
```

```python
from geopy.distance import geodesic

temp_lst = []
for row in coords['Координаты']:
    if geodesic(row.replace('[', '').replace(']', ''), '55.0415, 82.9346').meters < 100000:
        temp_lst.append(row)
```

```python
with open('temp.txt', 'w', encoding='utf8') as f:
    for i in temp_lst:
        f.writelines(i+',')
```
