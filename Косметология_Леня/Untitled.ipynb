```python
import pandas as pd
import os
```

```python
data = pd.read_excel('33cf026a-904b-4737-ade1-a5dfc4aa2ebe.xlsx')
```

```python
data = data[(data['Возраст'].isin(range(18, 46))) & (~data['Электронный адрес Пациента'].isnull())]
```

```python
codes = pd.read_excel('Копия Косметология.xls', dtype={'Код услуги' : str})['Код услуги'].to_list()
data1 = data[data['Код услуги'].isin(codes)]

```

```python
data1 = data1.drop_duplicates(subset=['Номер карты']).reset_index(drop=True)
data1
```

```python
data2 = data[data['Код МКБ-10 из списка ЭПИД'] == 'I87.8'].reset_index(drop=True)
data2 = data2.drop_duplicates(subset=['Номер карты'])
```

```python
data1.to_excel('Пациенты у косметолога.xlsx')
data2.to_excel('Пациенты с кодом I87.8.xlsx')
```
