```python
import pandas as pd
import datetime
import re
```

```python
data = pd.read_excel('Анестезия_акция.xlsx')
```

```python
data = data[data['Sum([Количество])'] > 0]
```

```python
data['Период'] = data['Дата приема'].dt.to_period('M')
```

```python
data.groupby(['Период', 'Наименование услуги'], as_index=False).agg({'Sum([Всего начислено])':sum, 'Sum([Количество])':'sum', 'Номер карты':'nunique'}).to_excel('Анестезия_акция_анализ.xlsx')
```
