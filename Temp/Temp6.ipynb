```python
def bingo(x):
    x = list(str(x))
    if all([int(i) % 2 for i in x]) and all([x.count(i) == 1 for i in x]):
        return True
    else:
        return False
```

```python
round(len([j for j in range(1000, 10000) if bingo(j) ]) / len(range(1000, 10000)), 3)
```

```python
def bingo(x):
    x = list(str(x))
    if all([x.count(i) == 1 for i in x]) and [int(i) for i in x] == sorted([int(i) for i in x], reverse=True):
        return True
    else:
        return False
```

```python
round(len([j for j in range(1000, 10000) if bingo(j) ]) / len(range(1000, 10000)), 5)
```

```python
def bingo(x):
    x = list(str(x))
    if all([x.count(i) == 1 for i in x]) and sum([int(x[i]) - int(x[i+1]) for i in range(3)]) == 3:
        return True
    else:
        return False
```

```python
round(len([j for j in range(1000, 10000) if bingo(j) ]) / len(range(1000, 10000)), 5)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a' : np.random.randn(6),
                  'b' : np.random.randn(6),
                  'key' : ['one', 'one', 'one', 'two', 'two', 'two']})

def aggr(arr):
    return list(arr)[-1]

df.groupby('key').agg(aggr)
```

```python
import pandas as pd
import numpy as np
from collections import Counter

np.random.seed(42)

df = pd.DataFrame({'a' : np.random.randint(1, 100, 100),
                  'b' : np.random.randint(1, 100, 100),
                  'key' : ['one'] * 50 + ['two'] * 50})

def aggr(arr):
    
    return Counter(arr).most_common(5)

df.groupby('key', as_index=False).agg(aggr)
```
