```python
import random
import numpy as np

'''Лента длиной L сантиметров разрезается на 2 куска в случайном месте.

Найдите вероятность того, что один из кусков имеет длину не менее M сантиметров.

На вход подаётся 2 числа: L, M

Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.'''


L = 100
M = 80

arr = np.arange(0.0001, L, 0.0001)
result = []

for _ in range(100000):
    x = random.choice(arr)
    lst = [x, L-x]
    if any([i >= M for i in lst]):
        result.append(1)
    else:
        result.append(0)    
        
print(str(round(sum(result) / len(result) * 100, 1)) + '%')        
    
```

```python
import random
import numpy as np

'''Лента длиной L сантиметров разрезается на N кусков в случайных местах (т.е. делают N-1 разрез).

Найдите вероятность того, что один из кусков имеет длину не менее M сантиметров.

На вход подаётся 3 числа: L, N, M

Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.'''


L = 100
N = 2
M = 80


result = []
lst = []

arr = np.arange(0.01, L, 0.01)
for _ in range(100000):
    lst.append(0)
    for j in range(N - 1):
        x = random.choice(arr)
        if x in lst:
            while x in lst:
                x = random.choice(arr)        
        lst.append(x)   
    lst.sort()
    lst.append(L)
    result.append(lst.copy())
    lst.clear()

def line_len(lst):
    cnt = []
    for i, j in enumerate(lst):
        if i == 0:
            temp_var = j
            continue
        cnt.append(j - temp_var)
        temp_var = j
    return cnt    

       
        
result = list(map(line_len, result))

result = [int(any([j >= M for j in i])) for i in result]
        
print(str(round(sum(result) / len(result) * 100, 1)) + '%')        
    
```

```python
'''В трамвайном парке имеются N трамваев маршрута №1 и M трамваев маршрута №2. Какова вероятность того, что вторым по счету на линию выйдет трамвай маршрута №1?

На вход подаётся 2 числа: N, M

Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.'''

N = 15
M = 10

arr = ['N'] * N + ['M'] * M

result = []

for _ in range(100000):
    arr.remove(random.choice(arr))
    y = random.choice(arr)
    if y == 'N':
        result.append(1)
    else:
        result.append(0)
    arr = ['N'] * N + ['M'] * M    

print(str(round(sum(result) / len(result) * 100, 1)) + '%') 
```

```python
'''В трамвайном парке имеются N трамваев маршрута №1 и M трамваев маршрута №2. Какова вероятность того, что K-м по счету на линию выйдет трамвай маршрута №1?

На вход подаётся 3 числа: N, M, K

Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.'''

N = 15
M = 10
K = 2

arr = ['N'] * N + ['M'] * M

result = []

for _ in range(100000):
    for j in range(K - 1):
        arr.remove(random.choice(arr))
    y = random.choice(arr)
    if y == 'N':
        result.append(1)
    else:
        result.append(0)
    arr = ['N'] * N + ['M'] * M    

print(str(round(sum(result) / len(result) * 100, 1)) + '%') 

```

```python
import re

s = '''rgb(0%,50%,100%)
rgba(0,0,0,0)  
rgb(255,255,255) 
rgb(0,,0) 
rgb(-1,0,0)  
rgba(0,0,0,1.5) 
rgba(0,0,0,0.5)'''.split('\n')

def right_rgb(row):
    row = row.strip()
    if row.split('(')[0] not in ('rgb', 'rgba'):
        return False
    if row[:4] == 'rgba':
        if '' in row.split(','):
            return False
        if len(row.split(',')) != 4:
            return False
        temp_s = row[5:-1].split(',')
        for element in temp_s[:-1]:
            if element[-1] != '%':
                if int(element) not in range(256):
                    return False
            else:
                if int(element[:-1]) not in range(101):
                    return False
        if float(temp_s[-1]) < 0 or float(temp_s[-1]) > 1:
            return False
    else:
        if len(row.split(',')) != 3:
            return False
        if '' in row.split(','):
            return False        
        temp_s = row[4:-1].split(',')
        for element in temp_s:
            if element[-1] != '%':
                if int(element) not in range(256):
                    return False
            else:
                if int(element[:-1]) not in range(101):
                    return False
    return True


r = '''rgb(0%,50%,100%) ---> True  
rgba(0,0,0,0)    ---> True  
rgb(255,255,255) ---> True  
rgb(0,,0)        ---> False  
rgb(-1,0,0)      ---> False  
rgba(0,0,0,1.5)  ---> False  
rgba(0,0,0,0.5)  ---> True'''

for line in s:
    print(right_rgb(line))
                
```
