```python
from datetime import date, datetime, time, timedelta
import pandas as pd

# создадим словарь из филиалов и дней недели, в который запишем начало и конец интервала для каждого из потоков
# всего 3 потока: "Здоровый ребенок", "Условно чистый поток", "Грязный поток"

filials = '''Б.С. 45
КИМа 28
Мар. 48
Брянцева 13
Гаврил.11
Маршака 4
Сик. 7к2
Колом. 28
Богатыр 49к1
Шаврова 26
Щерб.11
Колл. 4к1
Ленская 17к1
Гарьк. 15к3
Гаст. 22
Мор. 3к2
Ст.П. 39А
'''.splitlines()
days_of_week = '''пн
вт
ср
чт
пт
сб
вс
'''.splitlines()
dct = {}

for key in filials:
    dct.setdefault(key, {})
    for day in days_of_week:
        dct[key].setdefault(day, [])
```

```python
# отфитруем все лишнее
data = data[data['Неделя'] == 30]

data = data[['Филиал', 'День недели', 'Sum([Здоровый ребенок ИСПОЛНИТЕЛЬ])', 'Sum([Условно ЧИСТЫЙ ИСПОЛНИТЕЛЬ])', 'Sum([Условно ГРЯЗНЫЙ ИСПОЛНИТЕЛЬ])', 'Интервал приема Исполнителя']]
data = data[data['Филиал'].isin(dct.keys())]

data = data.reset_index(drop=True)
```

```python
# получим начало и конец проверяемого интервала
def razb(x):
    return [time(*map(int, i.split(':'))) for i in x.split(' - ')]

data['Для проверки'] = data['Интервал приема Исполнителя'].apply(razb)
```

```python
# функция для поминутного сравнения и получения нахождения времени в одном из потоков
# вернет список в котором каждое значение - кол-во минут из интервала в соответствующем потоке

def count_minutes(row):
    first = row['Для проверки'][0]
    second = row['Для проверки'][1]
    utro = dct[row['Филиал']][row['День недели']][0]
    den = dct[row['Филиал']][row['День недели']][1]
    vecher = dct[row['Филиал']][row['День недели']][2]
    x = 0
    cnt = [0,0,0]
    while first != second:
        x += 1
        dt = datetime.combine(date(2020, 7, 29), first) + timedelta(minutes=1)
        first = dt.time()
        if first > utro[0] and first <= utro[1]:
            cnt[0] += 1
        elif first > den[0] and first <= den[1]:
            cnt[1] += 1
        else:
            cnt[2] += 1
    return cnt        
    
```

```python
data['zdor'], data['chist'], data['gryzn'] = zip(*data.apply(count_minutes, axis=1))
```

```python
def minutes_from_lst(x):
    return time(*divmod(x, 60))
```
