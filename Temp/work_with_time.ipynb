```python
from datetime import date, datetime, time, timedelta
import pandas as pd

# создадим словарь из филиалов и дней недели, в который запишем начало и конец интервала для каждого из потоков
# всего 3 потока: "Здоровый ребенок", "Условно чистый поток", "Грязный поток"

filials = '''Б.С. 45
КИМа 28
Мар. 48
Брянцева 13
Гаврил.11
Маршака 4
Сик. 7к2
Колом. 28
Богатыр 49к1
Шаврова 26
Щерб.11
Колл. 4к1
Ленская 17к1
Гарьк. 15к3
Гаст. 22
Мор. 3к2
Ст.П. 39А
'''.splitlines()
days_of_week = '''пн
вт
ср
чт
пт
сб
вс
'''.splitlines()
dct = {}

for key in filials:
    dct.setdefault(key, {})
    for day in days_of_week:
        dct[key].setdefault(day, [])
```

```python
# начинаем добавлять объекты time для начало и конца каждого интервала по филиалам и дням недели

keys_first = ['Б.С. 45', 'КИМа 28', 'Колом. 28', 'Ленская 17к1', 'Ст.П. 39А']

for key in keys_first:
    for day in dct[key]:
        dct[key][day] = [(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]

for day in dct['Мар. 48']:
    dct['Мар. 48'][day] = [(time(0),time(0)), (time(8), time(20)), (time(0), time(0))]

keys_first = ['Гаврил.11', 'Маршака 4', 'Сик. 7к2']

for key in keys_first:
    for day in dct[key]:
        if day in ('ср', 'сб'):
            dct[key][day] = [(time(8),time(14, 30)), (time(14, 30), time(18)), (time(18), time(22))]
        else:
            dct[key][day] = [(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]

for day in dct['Богатыр 49к1']:
    dct['Богатыр 49к1'][day] = [(time(8),time(14, 30)), (time(14, 30), time(20)), (time(0), time(0))]

keys_first = ['Брянцева 13', 'Колл. 4к1', 'Гаст. 22']

for key in keys_first:
    for day in dct[key]:
        if day in ('чт', 'вс'):
            dct[key][day] = [(time(8),time(14, 30)), (time(14, 30), time(18)), (time(18), time(22))]
        else:
            dct[key][day] = [(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]

for day in dct['Щерб.11']:
    if day in ('чт', 'вс'):
        dct['Щерб.11'][day] = [(time(8),time(14, 30)), (time(14,30), time(20)), (time(0),time(0))]
    else:
        dct['Щерб.11'][day] = [(time(0),time(0)), (time(8), time(20)), (time(0),time(0))]

for day in dct['Шаврова 26']:
    if day in ('вт', 'пт'):
        dct['Шаврова 26'][day] = [(time(8),time(14, 30)), (time(14, 30), time(18)), (time(18), time(22))]
    else:
        dct['Шаврова 26'][day] =  [(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]

for day in dct['Гарьк. 15к3']:
    if day in ('ср', 'вс'):
        dct['Гарьк. 15к3'][day] = [(time(8),time(14, 30)), (time(14, 30), time(18)), (time(18), time(22))]
    else:
        dct['Гарьк. 15к3'][day] = [(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]

for day in dct['Мор. 3к2']:
    if day in ('ср', 'сб'):
        dct['Мор. 3к2'][day] = [(time(8),time(14, 30)), (time(14, 30), time(18)), (time(18), time(22))]
    else:
        dct['Мор. 3к2'][day] =[(time(0),time(0)), (time(8), time(18)), (time(18), time(22))]
```

```python
# получим таблицу в которой будем проверять время

data = pd.read_excel(r'R:\AmanovRA\Проверка_потоки.xlsx', sheet_name='Sheet1')
```

```python
# отфитруем все лишнее
data = data[data['Неделя'] == 30]

data = data[['Филиал', 'День недели', 'Sum([Здоровый ребенок ИСПОЛНИТЕЛЬ])', 'Sum([Условно ЧИСТЫЙ ИСПОЛНИТЕЛЬ])', 'Sum([Условно ГРЯЗНЫЙ ИСПОЛНИТЕЛЬ])', 'Интервал приема Исполнителя']]
data = data[data['Филиал'].isin(dct.keys())]

data = data.reset_index(drop=True)
```

```python
# получим начало и конец проверяемого интервала
def razb(x):
    return [time(*map(int, i.split(':'))) for i in x.split(' - ')]

data['Для проверки'] = data['Интервал приема Исполнителя'].apply(razb)
```

```python
# функция для поминутного сравнения и получения нахождения времени в одном из потоков
# вернет список в котором каждое значение - кол-во минут из интервала в соответствующем потоке

def count_minutes(row):
    first = row['Для проверки'][0]
    second = row['Для проверки'][1]
    utro = dct[row['Филиал']][row['День недели']][0]
    den = dct[row['Филиал']][row['День недели']][1]
    vecher = dct[row['Филиал']][row['День недели']][2]
    x = 0
    cnt = [0,0,0]
    while first != second:
        x += 1
        dt = datetime.combine(date(2020, 7, 29), first) + timedelta(minutes=1)
        first = dt.time()
        if first > utro[0] and first <= utro[1]:
            cnt[0] += 1
        elif first > den[0] and first <= den[1]:
            cnt[1] += 1
        else:
            cnt[2] += 1
    return cnt      
```

```python
data['zdor'], data['chist'], data['gryzn'] = zip(*data.apply(count_minutes, axis=1))
```

```python
def minutes_from_lst(x):
    return time(*divmod(x, 60))
```
