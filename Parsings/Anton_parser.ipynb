```python
data = pd.read_excel(r'C:\Users\AmanovRA\Anaconda3\Work-work\Parsings\Forum.xlsx')


arr = {}


for row in data['откуда']:
    arr.setdefault(row, [])
    try:
        resp = urlopen(row) # скачиваем файл
        html = resp.read().decode('utf8') # считываем содержимое
        soup = BeautifulSoup(html, 'html.parser') # делаем суп
        for link in soup.find_all(['a', 'span']):
            for attr in link.attrs:      
                if 'mc21.ru' in link.get(attr):
                    arr[row].append((link.text, link.get(attr)))
    except HTTPError as e:
        arr[row].append('Ошибка: ' + str(e.code))
        continue
    except UnicodeDecodeError:
        soup = BeautifulSoup(urlopen(row),"html.parser",from_encoding="windows-1251")
        for link in soup.find_all(['a', 'span']):
            for attr in link.attrs:      
                if 'mc21.ru' in link.get(attr):
                    arr[row].append((link.text, link.get(attr)))        
    except Exception as e:
        arr[row].append('Ошибка: ' + str(e))
    if len(arr[row]) == 0:
        arr[row].append('Нет переходов')
```

```python
data['переходы'] = arr.values()
```
