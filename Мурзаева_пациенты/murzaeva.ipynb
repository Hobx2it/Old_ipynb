```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_excel('Мурзаева И.Ю. от 18.09.xlsx', sheet_name='Исходник')
```

```python
data.drop(521, inplace=True)
```

```python
data['№ карты'] = data['№ карты'].astype(int)
```

```python
first_visit = pd.read_excel('Пациенты_с_приемами.xlsx')
```

```python
def to_int(x):
    try:
        x = int(x)
        return x
    except:
        return x


first_visit['Номер карты'] = first_visit['Номер карты'].apply(to_int)
```

```python
data = data.merge(first_visit[['Номер карты', 'Дата первого посещения Пациента']], how='left', left_on='№ карты', right_on='Номер карты')
```

```python
data['Дата первого посещения Пациента'] = data['Дата первого посещения Пациента'].fillna('Новый пациент')
```

```python
data.drop('Номер карты', axis=1, inplace=True)
```

```python
services = pd.read_excel('Услуги_с_февраля.xlsx')
```

```python
services['Номер карты'] = services['Номер карты'].apply(to_int)
```

```python
data = data.merge(services, how='left', left_on='№ карты', right_on='Номер карты')
```

```python
data.sort_values(by=['№ карты'])
```

```python
data['Разница дней'] = data['Дата записи'] - data['Дата приема_y']
```

```python
data['Разница дней'] = data['Разница дней'].apply(lambda x: x.days)
```

```python
def count_month(x):
    if 0 <= x <= 30:
        return "1 месяц"
    elif 31 <= x <= 60:
        return "2 месяца"
    elif x >= 61:
        return "Более 2 месяцев"
    elif x < 0:
        return "Есть прием после дня записи"
    else:
        return "Не было приемов"
    
data['Сроки приема'] = data['Разница дней'].apply(count_month)    
```

```python
data.drop_duplicates(subset=['Пациент_x', 'Доктор'], inplace=True)
```

```python
data = data.reset_index(drop=True)
```

```python
data = data.fillna('')
```

```python
data['№ карты'].nunique()
```

```python
data.to_excel('Выгрузка_пациенты_из_записи_обработано.xlsx', index=None)
```

```python
group_data = data.groupby(by=['Сроки приема', 'Доктор', 'Отделение'], as_index=False).agg({'№ карты' : 'count'})
```

```python
group_data.rename(columns={'№ карты' : 'Количесмтво пациентов'}, inplace=True)
```

```python
group_data.rename(columns={'Количесмтво пациентов' : 'Количество пациентов'}, inplace=True)
```

```python
group_data.groupby(by='Сроки приема').agg({'Количество пациентов':'sum'}).plot(kind='pie', subplots=True)
```
