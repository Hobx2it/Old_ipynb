```python
"""Задание 3
По данным портала открытых данных Москвы определите количество дней, когда освещение включено 12 часов или более.
"""
# Работаем с apidata.mos.ru
key = '38499f0c97b025b14c841205629827bd' # уникальный ключ пользователя в систему
dataSetNumber = '3288' # у каждого набора данных этого api свой номер, это номер датасета с графиком отключения воды
cnt = urllib.request.urlopen('https://apidata.mos.ru/v1/datasets/'+dataSetNumber+'/count?api_key='+key).read()
rows = urllib.request.urlopen('https://apidata.mos.ru/v1/datasets/'+dataSetNumber+'/rows?api_key='+key+'&$top=500').read().decode('utf-8')
#print(rows, '\n')
lst = json.loads(rows)  # переводит в питоновский список
days = []

for day in lst:
    cells = day['Cells']
    duration = int(cells['DurationOfLighting'][:2])
    if duration >= 12:
        days.append(cells['Date'])

print(len(days))     
```

```python
"""Задание 4
Вам дан xlsx-файл с информацией о спортивных площадках в Москве. Вам необходимо сдать на проверку json-файл, в котором будет храниться один словарь, ключами которого будут административные округа (AdmArea), а значениями словари, в которых, в свою очередь, ключами будут названия районов (District), относящихся к этому административному округу, а значениями - списки адресов площадок (Address) в том порядке, в котором они встречались в исходном файле.

Ваш файл должен выглядеть примерно так:

{"Северо-Западный административный округ": {"район Строгино": ["улица Исаковского, дом 24, корпус 1", "Неманский проезд, дом 9"], "район Северное Тушино": ["улица Свободы, дом 56", "улица Свободы, дом 56", "улица Свободы, дом 56", "улица Свободы, дом 56"], "район Покровское-Стрешнево": ["Иваньковское шоссе, дом 6"], "район Щукино": ["Сосновая улица, дом 3, строение 2"]}, ...}
"""
key = '38499f0c97b025b14c841205629827bd' # уникальный ключ пользователя в систему
dataSetNumber = '893' # у каждого набора данных этого api свой номер, это номер датасета с графиком отключения воды
cnt = urllib.request.urlopen('https://apidata.mos.ru/v1/datasets/'+dataSetNumber+'/count?api_key='+key).read()
rows = urllib.request.urlopen('https://apidata.mos.ru/v1/datasets/'+dataSetNumber+'/rows?api_key='+key+'&$top=500').read().decode('utf-8')
#print(rows, '\n')
lst = json.loads(rows)  # переводит в питоновский список
i=0
while i < int(cnt):
    i += 500
    rows = urllib.request.urlopen('https://apidata.mos.ru/v1/datasets/'+dataSetNumber+'/rows?api_key='+key+'&$top=500'+'&$skip'+str(i)).read().decode('utf-8')
    lst.extend(json.loads(rows))
    print(i)
```

```python
dct ={}
for i in lst:
    cells = i['Cells']
    AdmArea = cells['AdmArea']
    District = cells['District']
    Address = cells['Address']
    dct.setdefault(AdmArea, {})
    dct[AdmArea].setdefault(District, [])
    if Address not in dct[AdmArea][District]:
        dct[AdmArea][District].append(Address)
print(dct) 

fout = open('json1', 'w', encoding = 'utf8')
print(dct, file = fout)
fout.close()
    
```
