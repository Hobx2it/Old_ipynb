```python
import pandas as pd
import os
```

```python
zoon_spb = pd.DataFrame()

os.chdir(r'C:\Users\AmanovRA\Текущее\Zoon_1_квартал\Зун 1 квартал Спб + Нск\Спб')
for file in os.listdir('.'):
    temp_df = pd.DataFrame()
    xl = pd.ExcelFile(file)
    for sheet in xl.sheet_names:
        temp_df = temp_df.append(pd.read_excel(file, sheet_name=sheet), ignore_index=True)
    temp_df['Филиал'] = [file[:file.find('.')] for i in range(temp_df.shape[0])]
    zoon_spb = zoon_spb.append(temp_df, ignore_index=True)
zoon_spb['Телефон'] = zoon_spb['Телефон'].apply(int).apply(str)    
        
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
%matplotlib inline

au_telefons = pd.read_excel(r'C:\Users\AmanovRA\Текущее\Zoon_1_квартал\АУ_по_июнь.xlsx')

dom, mobil, work = au_telefons['Телефон домашний'].tolist(), au_telefons['Телефон мобильный'].tolist(), au_telefons['Телефон рабочий'].tolist()

pattern = r'\(\d{2}\)'
pattern2 = r'[а-яА-Яa-zA-Z]{1}'

for key in (dom, mobil, work):
    for i, j in enumerate(key):
        if type(j) == str:
            key[i] = ''.join([x if x not in ',;!:.\\/+' else ' ' for x in j])        
            key[i] = key[i].lstrip(' ')
            key[i] = key[i].rstrip(' ')
            if re.search(pattern2, key[i]):
                key[i] = re.sub(pattern2, '', key[i])
            key[i] = ''.join([x for x in key[i] if x not in '()-'])  
            
for key in (dom, mobil, work):
    for i, j in enumerate(key):
        if type(j) == str:
            if len(j) == 7:
                key[i] = '7812' + key[i]
            elif len(j) == 11 and j[0] == '8':
                key[i] = '7' + j[1:]
            elif len(j) == 10:
                key[i] = '7' + j
                
def concate_lists(*args) -> list:   # соединим все номера в одну строку чтобы проще было искать
    temp_lst = []
    for i in range(len(args[0])):
        temp_lst.append(' '.join([str(x[i]) for x in args]))
    return temp_lst
            
lst_search = concate_lists(dom, mobil, work)                 
```

```python
lst_with_coincidenc_telefon = []
lst_with_coincidenc_filial = []
for index, row in enumerate(lst_search):
    lst_with_coincidenc_telefon.append([])
    lst_with_coincidenc_filial.append([])
    for i, j in enumerate(spb_telefons['Телефон']):
        if str(j) in row:
            lst_with_coincidenc_telefon[index].append(str(j))
            lst_with_coincidenc_filial[index].append(spb_telefons['Филиал'][i])
            
au_telefons['Zoon-телефон'], au_telefons['Zoon-филиал'] = lst_with_coincidenc_telefon, lst_with_coincidenc_filial            
            
def join_list(lst):
    return ', '.join([str(i) for i in lst])

au_telefons['Zoon-телефон'], au_telefons['Zoon-филиал'] = au_telefons['Zoon-телефон'].apply(join_list), au_telefons['Zoon-филиал'].apply(join_list)
```

```python
au_telefons.to_excel(r'C:\Users\AmanovRA\Текущее\Zoon_1_квартал\СПБ_ZOON_по_июнь.xlsx')
```
