```python
import pandas as pd
import datetime
```

```python
data = pd.read_excel('МатвеевДВ.xlsx')
```

```python
data.head()
```

```python
data['Период'] = data['Дата приема'].apply(lambda x: (pd.to_datetime(x).date(), pd.to_datetime(x).date() - datetime.timedelta(30)))
```

```python
arr = []
for i in data.index:
    if data['Категория крупно'][i] != 'операции':
        arr.append('')
        continue
    else:
        number_card = data['Номер карты'][i]
        finish, start = data['Период'][i]
        temp_data = data[(data['Номер карты'] == number_card) & (data['Категория крупно'].str.contains('консультац'))]['Дата приема'].to_list()
        temp_arr = [pd.to_datetime(j).date().strftime('%d/%m/%Y') for j in temp_data if pd.to_datetime(j).date() >= start and j <= finish]
        if len(temp_arr) == 0:
            arr.append('')
        else:    
            arr.append(temp_arr)
```

```python
arr
```

```python
data['Вердикт'] = arr
```

```python
data.to_excel('Вердикт.xlsx')
```
