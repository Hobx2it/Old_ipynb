```python
"""Задача 9 
Напишите программу, которая будет разрезать большую прямоугольную область на N×N одинаковых прямоугольных областей. Области задаются четырьмя координатами: минимальной широтой, минимальной долготой, максимальной широтой, максимальной долготой.

При выводе области должны быть упорядочены по возрастанию минимальной широты, а в случае равных широт - по возрастанию минимальной долготы.

Гарантируется, что все числа во входных данных положительны. 

Sample Input:

41.173 77.23 42.17 79.004
2
Sample Output:

41.173 77.23 41.6715 78.117
41.173 78.117 41.6715 79.004
41.6715 77.23 42.17 78.117
41.6715 78.117 42.17 79.004"""

arr = [float(i) for i in '41.173 77.23 42.17 79.004'.split()]
n = 2

minLat = arr[0]
maxLat = arr[2]
minLon = arr[1]
maxLon = arr[3]

dLat = (maxLat - minLat) / n
dLon = (maxLon - minLon) / n

print(dLat, dLon)

for i in range(n):
    for j in range(n):
        nMinLat = minLat + dLat * i
        nMaxLat = minLat + dLat * (i + 1)
        nMinLon = minLon + dLon * j
        nMaxLon = minLon + dLon * (j + 1)
        print(nMinLat, nMinLon, nMaxLat, nMaxLon)
        
```

```python
"""Задача 15
Вам дана область карты https://stepik.org/media/attachments/lesson/266078/mapcity.osm

Пройдите по всем way в этой области, выделите среди них замкнутые (те, у которых совпадает ссылка на первый и последний node), среди всех замкнутых выделите те, у которых установлен tag с ключом building и произвольным значением.

Запомните id для way и список кортежей, содержащих координаты (широту и долготу) всех node, входящих в этот way. 

Вам предложена функция, которая определяет нечто похожее на площадь замкнутой ломаной.
"""
import math

resp = urlopen('https://stepik.org/media/attachments/lesson/266078/mapcity.osm')
xml = resp.read().decode('utf8') # считываем содержимое
soup = BeautifulSoup(xml, 'xml')

ways =[]
for way in soup.find_all('way'):
    nds =[]
    for nd in way('nd'):
        nds.append(nd.get('ref'))
        if len(nds) > 1 and nds[0] == nds[-1]:
            for tag in way('tag'):
                if tag['k'] == 'building':
                    ways.append((way.get('id'), nds))

ways = list(map(list, ways))

def coordinates(ref):
    for node in soup.find_all('node'):
        if node.get('id') == ref:
            return (float(node.get('lat')), float(node.get('lon')))
ways2 = []  

for i in ways:
    ways2.append([])
    ways2[ways.index(i)].append(ways[ways.index(i)][0])
    ways2[ways.index(i)].append([])
    for j in i[1]:
        ways2[ways.index(i)][1].append(coordinates(j))
        


def getsqr(coordlist):
    baselat = coordlist[0][0]
    baselon = coordlist[0][1]
    degreelen = 111300
    newcoord = []
    for now in coordlist:
        newcoord.append(((now[0] - baselat) * degreelen, (now[1] - baselon) * degreelen * math.sin(baselat)))
    sqr = 0
    for i in range(len(newcoord) - 1):
        sqr += newcoord[i][0] * newcoord[i + 1][1] - newcoord[i + 1][0] * newcoord[i][1]
    sqr += newcoord[-1][0] * newcoord[0][1] - newcoord[0][0] * newcoord[-1][1]
    return abs(sqr)

maxSqr = 0
maxId = ''

for i in ways2:
    if getsqr(i[1]) > maxSqr:
        maxSqr = getsqr(i[1])
        maxId = i[0]
        
print(maxId)        

    
    
    
```
