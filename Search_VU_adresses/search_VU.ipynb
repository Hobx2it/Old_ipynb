```python
import requests
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
import json
import openpyxl
import pandas as pd

api_keys = iter(['0968d0a9-5bca-44f0-861f-bf9beaf2a82a', 'b996ed8b-dc1b-4dee-8687-bbb4b26dd313'])

url = 'https://geocode-maps.yandex.ru/1.x?geocode='


def addr(b, api_key):
    a = b.replace(' ', '')
    a = str(a.encode('utf8'))
    i = a[2]
    a2 = ''.join(a.split(i))
    a3 = a2.replace('x', '%')
    resp = urlopen(url + a3.upper() + '&apikey=' + api_key + '&format=json')
    rows = resp.read().decode('utf8') # считываем содержимое
    lst = json.loads(rows)
    #coords = lst['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['Point']['pos'].split()
    return lst
```

```python
data_vu_adresses = pd.read_excel(r'C:\Users\AmanovRA\Anaconda3\Work-work\Search_VU_adresses\SPB\Адрес_для_поиска_СПБ_2.xlsx')
data_vu_adresses = data_vu_adresses.fillna('')
```

```python
adress_for_search = []
for i in data_vu_adresses.index:
    temp_house_liter_korp_arr = []
    for j in ('Дом1', 'Корп1', 'Литера1'):
        if data_vu_adresses.loc[i, j] and sum(1 for x in str(data_vu_adresses.loc[i, j]) if x.isdigit() or x.isalpha()) > 0:
            if j != 'Литера1':
                temp_house_liter_korp_arr.append(f', {j[:-1]} {data_vu_adresses.loc[i, j]}')
            else:
                if str(data_vu_adresses.loc[i, j]).isnumeric():
                    temp_house_liter_korp_arr.append(f', Строение {data_vu_adresses.loc[i, j]}')
                else:
                    temp_house_liter_korp_arr.append(f', {j[:-1]} {data_vu_adresses.loc[i, j]}')
    adress_for_search.append(data_vu_adresses.loc[i, 'Адрес'] + ''.join(temp_house_liter_korp_arr).strip()) 
```

```python
data_vu_adresses['Адрес_для_поиска'] = adress_for_search
data_vu_adresses
```

```python
%%time

json_objects = []
adresses = []

cnt = 0
flag  = False

for adress in data_vu_adresses['Адрес_для_поиска']:
    adresses.append(adress)
    cnt += 1
    if cnt > 19500 or not flag:
        key = next(api_keys)
        cnt = 1
        flag = True        
    try:
        json_objects.append(addr(adress, key))    
    except Exception as err:
        json_objects.append('Ошибка ' + str(err))
        continue
        
json_objects   
```

```python
data_vu_adresses['Python_JSON'] = json_objects
```

```python
def coords_from_json(s):
    try:
        return ' '.join(s['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['Point']['pos'].split()[::-1])
    except:
        return ''
```

```python
data_vu_adresses['Coord'] = data_vu_adresses['Python_JSON'].apply(coords_from_json)
```

```python
json_true = []
for i in json_objects:
    json_true.append(json.dumps(i, ensure_ascii=False))
```

```python
data_vu_adresses['True_JSON'] = json_true
```

```python
data_vu_adresses = data_vu_adresses.drop('Python_JSON', axis=1)
```
