```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
vu_telefons = pd.read_excel(r'C:\Users\AmanovRA\Текущее\Лиды_апрель_май\ВУ_по_июнь.xlsx')
vu_telefons
```

```python
phones = vu_telefons['Телефон'].tolist()
```

```python
pattern = r'\(\d{2}\)'
pattern2 = r'[а-яА-Яa-zA-Z]{1}'

for i, j in enumerate(phones):
    if len([x for x in j if x.isdigit()]) == 10:
        phones[i] = '7' + ''.join([x for x in j if x.isdigit()])    
    if len([x for x in j if x.isdigit()]) == 11:
        temp_arr = ''.join([x for x in j if x.isdigit()])
        if temp_arr[0] == '8':
            phones[i] = '7' + temp_arr[1:]
        elif temp_arr[0] == '7':    
            phones[i] = temp_arr

for i, j in enumerate(phones):
    if re.search(pattern, j):
        if '(61)' in j:
            phones[i] = phones[i].replace('(61)', ', 321-61-61, ')
        elif  '(39)' in j:   
            phones[i] = phones[i].replace('(39)', ', 305-73-39, ')
        elif '(73)' in j:
            phones[i] = phones[i].replace('(73)', '')
            phones[i] += ' , 558-73-09 '
        elif ('21') in j:
            phones[i] = phones[i].replace('(21)', '(921)')
    phones[i] = ''.join([x if x not in ',;!:.\\/+' else ' ' for x in j])        
    phones[i] = phones[i].lstrip(' ')
    phones[i] = phones[i].rstrip(' ')
    if re.search(pattern2, phones[i]):
        phones[i] = re.sub(pattern2, '', phones[i])
    phones[i] = ''.join([x for x in phones[i] if x not in '()-']) 
```

```python
phones2 = phones.copy()
phones2
```

```python
for i, j in enumerate(phones2):
    phones2[i] = j.split()
```

```python
from functools import reduce
for i, j in enumerate(phones2):
    try:
        phones2[i].append(reduce(lambda x, y: x + y, [z for z in j if len(z) < 7]))
    except:
        continue
```

```python
for i, j in enumerate(phones2):
    for x, y in enumerate(j):
        if x != len(j) - 1:
            if len(y) == 4 and len(j[x + 1]) == 7 and y[:2] in ('79', '89'):
                    phones2[i].append(y + j[x + 1])
            elif y in ('7', '8') and len(j[x + 1]) == 10:
                phones2[i].append(y + j[x + 1])
```

```python
for i, j in enumerate(phones2):
    phones2[i] = [x for x in phones2[i] if len(x) >= 7]
```

```python
for i, j in enumerate(phones2):
    for x, y in enumerate(j):
        if len(y) == 7:
            phones2[i][x] = '7812' + phones2[i][x]
        elif len(y) == 11 and y[0] == '8':
            phones2[i][x] = '7' + y[1:]
        elif len(y) == 10:
            phones2[i][x] = '7' + y
```

```python
vu_telefons['Чистые_телефоны'] = phones2
vu_telefons['Чистые_телефоны']
```

```python
april, may = pd.read_excel(r'C:\Users\AmanovRA\Текущее\Лиды_апрель_май\Лиды_апрель_май_АУ.xlsx', sheet_name='Апрель'), pd.read_excel(r'C:\Users\AmanovRA\Текущее\Лиды_апрель_май\Лиды_апрель_май_АУ.xlsx', sheet_name='Май')
```

```python
april['Клиент: телефон'] = april['Клиент: телефон'].apply(str)
may['Клиент: телефон'] = may['Клиент: телефон'].apply(str)
```

```python
lst_with_coincidenc_april = []
for index, row in enumerate(phones2):
    lst_with_coincidenc_april.append([])
    for i in april['Клиент: телефон']:
        if str(i) in row:
            lst_with_coincidenc_april[index].append(str(i))
```

```python
lst_with_coincidenc_may = []
for index, row in enumerate(phones2):
    lst_with_coincidenc_may.append([])
    for i in may['Клиент: телефон']:
        if str(i) in row:
            lst_with_coincidenc_may[index].append(str(i))
```

```python
vu_telefons['Апрель'], vu_telefons['Май'] = lst_with_coincidenc_april, lst_with_coincidenc_may
```

```python
def join_list(lst):
    return ', '.join([str(i) for i in lst])

vu_telefons['Апрель'], vu_telefons['Май'] = vu_telefons['Апрель'].apply(join_list), vu_telefons['Май'].apply(join_list)
```
