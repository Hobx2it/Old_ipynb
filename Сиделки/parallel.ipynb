```python
import pandas as pd
```

```python
parallel = pd.read_excel('Параллельные_услуги.xlsx')
```

```python
sestr = parallel[parallel['Наименование услуги'].str.contains('сестринский пост')].reset_index(drop=True).groupby('Номер карты', as_index=False).agg({'Дата приема' : list}).rename(columns={'Дата приема' : 'Даты сестринских постов'})
```

```python
other_services = parallel[~parallel['Наименование услуги'].str.contains('сестринский пост')].reset_index(drop=True)
```

```python
other_services = other_services.merge(sestr)
```

```python
temp_lst = []
for index in other_services.index:
    if other_services['Дата приема'][index] in other_services['Даты сестринских постов'][index]:
        temp_lst.append(True)
    else:
        temp_lst.append(False)
        
other_services['Check_dates'] = temp_lst        
```

```python
final = other_services[other_services['Check_dates'] == True].reset_index(drop=True)
```

```python
final.to_excel('Услуги_вместе_с_постами.xlsx', index=None)
```
