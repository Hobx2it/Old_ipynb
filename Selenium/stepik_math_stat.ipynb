```python
from itertools import permutations
import random

row = [0,0,0,0,0,0,0,1]
```

```python
def du_matrix(row):
    matrixes = []
    while True:
        matrix = []
        for _ in range(10):
            temp_row = row.copy()
            random.shuffle(temp_row)
            matrix.append(temp_row)
        if matrix not in matrixes:
            matrixes.append(matrix)
            yield matrix
        
        
matrix_gen = du_matrix(row)

arr = [next(matrix_gen) for i in range(10)]

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

link = 'https://stepik.org/lesson/368500/step/8?unit=360397'

def matrix_to_web(matrix, browser, link):
    browser.implicitly_wait(5)
    browser.get(link)
    table = WebDriverWait(browser, 7).until(EC.visibility_of_element_located((By.CSS_SELECTOR, '.table-quiz__table')))
    for index, row in enumerate(matrix):
        row_index = index + 1
        col_index = row.index(1) + 1
        browser.find_element_by_css_selector(f'.table-quiz__table tbody tr:nth-child({row_index}) td:nth-child({col_index})').click()     
        
        
with webdriver.Chrome() as browser:
    matrix_to_web(du_matrix(row), browser, link)
    time.sleep(10)
    
    
    
```
